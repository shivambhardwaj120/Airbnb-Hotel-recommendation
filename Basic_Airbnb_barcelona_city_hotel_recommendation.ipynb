# Content based recommedation(by room name)

```python
df35=df33.sample(frac=0.4, replace=False, random_state=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer()
count_matrix = count.fit_transform(df35['neighborhood'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
```

```python
df33 = df33.reset_index()
indices = pd.Series(df35.index, index=df33['name'])
```

```python
def C_rec(name, cosine_sim=cosine_sim):
    idx = indices[name]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    room_indices = [i[0] for i in sim_scores]
    return df35[['name','price','overall_satisfaction']].iloc[room_indices]
```

```python
df_rec=C_rec('GORGEOUS EQUIPPED APTM WITH PARKING', cosine_sim2)
```

```python
df_rec=pd.DataFrame(df_rec)
```
